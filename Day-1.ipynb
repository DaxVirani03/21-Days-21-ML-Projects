```python
df['R'] = df['R'].map({'R':0,'M':1})
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr())
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
Y=df['R']
X=df.drop('R',axis=1)
```

```python
X_sc = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_sc,Y,test_size=.20,random_state=42)
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(accuracy_score(Y_test,y_pred))
```

```python
print(classification_report(Y_test, y_pred))
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
pca = PCA(n_components = 6)
X_pca = pca.fit_transform(X_sc)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_pca,Y,test_size=.20,random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(accuracy_score(Y_test,y_pred))
```

```python
accuracy_scores = []
n_features = X_train.shape[1]   # total number of features

for n in range(1, n_features + 1):
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)   # fit on training data
    X_test_pca = pca.transform(X_test)         # transform test data
    
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train_pca, Y_train)
    y_pred = model.predict(X_test_pca)
    
    acc = accuracy_score(Y_test, y_pred)
    accuracy_scores.append(acc)

# plot accuracy vs PCA components
plt.figure(figsize=(8,5))
plt.plot(range(1, n_features + 1), accuracy_scores, marker='o')
plt.xlabel("Number of PCA Components")
plt.ylabel("Accuracy Score")
plt.title("Accuracy vs PCA Components")
plt.grid(True)
plt.show()
```
